## Load data sources

````python
# Load data
import pandas as pd

# Load and keep only the relevant columns
df = pd.read_csv('/content/merged_results.csv')
df = df[['title', 'page_text', 'conversation_text', 'link']].dropna()
df
````

## Create a CrewAI agent

````python
!pip install crewai crewai_tools --quiet
from crewai import LLM, Agent, Task, Crew
import os

# ---- 1. LLM wrapper -------------------------------------------------
llm_crew = LLM(
    model="gemini/gemini-1.5-flash-latest",       # provider prefix
    api_key=api_key          # set GOOGLE_API_KEY env var
)

# # -- LLM that determines if the idea is worth investigating --
# 2. Define your single-call filter function
def filter_idea(text: str) -> str:
    prompt = (
        "Decide if this is a serious, effort-backed startup idea. "
        "Say YES only if there’s real implementation detail, code, benchmarks or traction. "
        "Otherwise say NO.\n\n"
        f"{text}"
    )
    resp = llm.invoke(prompt).content.strip().upper()
    return resp


# --- Agent (crew_research): define research agent ----
# Define Your Analyzer and Scorer Agents/Tasks
idea_analyzer = Agent(
    role="Startup Research Analyst",
    goal="Evaluate startup ideas for novelty, market fit, integration, and investment readiness",
    backstory=(
        "You're a detail-oriented analyst at a top VC firm. You analyze early-stage startup ideas using a set "
        "of strategic questions to understand their value proposition, problem, market, uniqueness, and team credibility."
    ),
    llm=llm_crew,
    tools=[Serpertool], #
    #step_callback= collect_raw, #collect logging
    verbose=True
)

scoring_agent = Agent(
    role="Investment Fit Evaluator",
    goal="Score each startup based on novelty, traction, team, market, and strategic fit",
    backstory=(
        "You're a senior VC partner who evaluates startup pitches by scoring key criteria like team quality, "
        "market size, defensibility, and timing. You look for high-signal opportunities for potential follow-ups."
    ),
    llm=llm_crew,
    tools=[], #Serpertool
    #step_callback= collect_raw, #collect logging
    verbose=True
)

# 1️⃣  embed {input} in each Task description
task_analysis = Task(
    description=(
        "Analyse the following startup post. If you need to confirm or enrich any point, call Serpertool:\n\n{input}\n\n"
        "- Who created it?\n"
        "- What problem does it solve (frequency / severity / industry)?\n"
        "- How does it work technically?\n"
        "- What’s unique vs existing solutions?\n"
        "- Pricing model & ease of integration\n"
        "- Target user (devs / enterprise / consumers)\n"
        "- When was it introduced?\n"
        "- Backers / funding info?\n"
        "Format ReAct-style: Thought / Action / Observation / Final Answer."
    ),
    expected_output="Structured markdown or JSON.",
    agent=idea_analyzer,
)

task_scoring = Task(
    description=(
        "Using the same post:\n\n{input}\n\n"
        "Score 1–5 on: Novelty, Market potential, Traction, Team, "
        "Integration, Strategic fit, Timing.\n"
        "Return JSON with the seven scores, the average, and a short justification."
    ),
    expected_output="JSON with 7 scores, average, justification.",
    agent=scoring_agent,
)


crew_research = Crew(
    agents=[idea_analyzer, scoring_agent],
    tasks=[task_analysis, task_scoring],
    verbose=False
)

def run_pipeline_on_row(row):
    idea_text = (
        f"Title: {row['title']}\n\n"
        f"Page Text:\n{row['page_text'][:600]}\n\n"
        f"Conversation:\n{row['conversation_text'][:600]}\n\n"
        f"Link: {row['link']}"
    )

    # simple filter (assume you already fixed this)
    # if filter_idea(idea_text) != "YES":
    #     return {"status": "SKIPPED"}

    results = crew_research.kickoff_for_each(inputs=[{"input": idea_text}])
    return results
````

## Run Agents for Analysis

````python
# The results below show the run AFTER the Serp search tool
demo_results = run_pipeline_on_row(df.iloc[7])
print(demo_results)
````

# Print out CrewAI analysis results

````python
import json, re, textwrap

def strip_json_fence(s: str) -> str:
    """
    Remove leading  ```
    or  ```json    and the closing ``` .
    Also works for strings that only start with 'json'.
    """
    s = s.strip()
    # remove the leading fence (if it exists)
    s = re.sub(r"^```json|^```|^json", "", s, flags=re.IGNORECASE).lstrip()
    # remove trailing fence
    s = re.sub(r"```$", "", s).rstrip()
    return s

def load_task_output(task_output) -> dict:
    """Take a TaskOutput object, return parsed JSON dict (handles fence stripping)."""
    return json.loads(strip_json_fence(task_output.raw))

def parse_demo_results(demo_results):
    """
    demo_results : list[ CrewOutput ]    (what you printed)
    Returns (analysis_dict, scoring_dict)
    """
    crew_output = demo_results[0]                      # only one element in your list
    # There are two TaskOutput objects in .tasks_output
    analysis_dict = load_task_output(crew_output.tasks_output[0])
    scoring_dict  = load_task_output(crew_output.tasks_output[1])
    return analysis_dict, scoring_dict

def pretty_analysis(d: dict) -> str:
    lines = [
        f"📌  **{d.get('Title','(no title)')}**",
        f"│  Creator: {d.get('Creator','?')}",
        f"│  Intro date: {d.get('Introduction_date','?')}",
        "│  Problem:",
        f"│    • {d['Problem']['description']}",
        f"│    • Freq: {d['Problem']['frequency']}",
        f"│    • Severity: {d['Problem']['severity']}",
        f"│    • Industry: {d['Problem']['industry']}",
        f"│  Technical solution: {d.get('Technical_solution')}",
        f"│  Uniqueness: {d.get('Uniqueness')}",
        f"│  Target user: {d.get('Target_user')}",
        f"│  Backers/Funding: {d.get('Backers_Funding')}",
    ]
    return "\n".join(lines)

def pretty_scores(d: dict) -> str:
    key_order = ["Novelty","MarketPotential","Traction","Team",
                 "Integration","StrategicFit","Timing","Average"]
    bullets = "\n".join([f"  • {k:15}: {d[k]}" for k in key_order if k in d])
    return textwrap.dedent(f"""\
        🎯 **Investment-fit scores**
        {bullets}
        ── Justification ──
        {d.get("Justification")}
    """)

# ---------- run it ----------
analysis, scoring = parse_demo_results(demo_results)
print(pretty_analysis(analysis))
print()
print(pretty_scores(scoring))
````
